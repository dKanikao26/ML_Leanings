```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(data.drop('Sleep Disorder' , axis = 1),data['Sleep Disorder'],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder(categories=[['Obese' , 'Overweight' , 'Normal' , 'Normal Weight']])
x_train['BMI Category'] = ordinal_encoder.fit_transform(x_train[['BMI Category']])
x_test['BMI Category'] = ordinal_encoder.transform(x_test[['BMI Category']])
```

now for nominal use one hot encoding(gender , occupation)

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False, drop = 'first')
x_train_encoded = one_hot_encoder.fit_transform(x_train[['Gender', 'Occupation']])
x_test_encoded = one_hot_encoder.transform(x_test[['Gender' , 'Occupation']])
```

```python
x_train_encoded = pd.DataFrame(x_train_encoded, columns=one_hot_encoder.get_feature_names_out(['Gender', 'Occupation']))
x_train_encoded
```

```python
x_train.drop(['Gender' , 'Occupation'] , axis = 1 , inplace = True)
x_test.drop(['Gender' , 'Occupation'] , axis = 1 , inplace = True)
```
